```python
import pandas as pd

dataset = pd.read_csv ("wine.data") #read data set
```

```python
dataset.columns = ['Class', 'Alcohol', 'Malic_acid', 
              'Ash', 'Pay_Alcalinity_of_ash', 
              'Magnesium', 'Total phenols'
             ,'Flavanoids', 'Nonflavanoid_phenols',
              'Proanthocyanins','Color intensity','Hue', 
              'OD280/OD315_of_diluted_wines',
             'Proline']
```

```python
import seaborn as sns;
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
heatplot = sns.heatmap(dataset.corr(), cmap='rainbow')
```

```python
label_class = dataset['Class'] 
dataset.drop(columns='Class', axis = 1, inplace=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset, label_class, test_size=0.2, random_state=24)
```
